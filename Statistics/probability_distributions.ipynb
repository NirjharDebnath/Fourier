## Importing The Libraries

```python
import numpy as np
import pandas as pd
import random
import sklearn
import sympy as sp
import matplotlib.pyplot as plt

plt.style.use("seaborn-v0_8-whitegrid")
```

## Unifrom Distribution

```python
a, b = 2, 4
x = np.linspace(2, 4, 1000)
def f(x):
    return [1/(b - a) for _ in range(len(x))]

y = f(x)
```

```python
from IPython.display import display, HTML
display(HTML("<style>div.output_png { display: flex; justify-content: center; }</style>"))

plt.figure(figsize=(10, 8), dpi=120)
plt.plot(x, y, label="f(x) = 1/(b - a)")
plt.axhline(0, color='black', linewidth=1, alpha=0.8)
plt.axvline(0, color='black', linewidth=1, alpha=0.8)
plt.axhline(y=0.5, xmin=0, xmax=1, color='blue', linestyle="--", alpha=0.4)
plt.fill_between(x, y, color='skyblue', alpha=0.4, label='Area under curve')
x_range = max(abs(a), abs(6))
plt.xlim(-x_range, x_range)
plt.ylim(-0.1, max(y) + 0.2)
plt.legend()
plt.grid(linestyle="--", alpha=0.4)
plt.show()
```
